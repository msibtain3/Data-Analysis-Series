```python
# importing relevant libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# loading the dataset
sales = pd.read_csv('sales_data.csv')
```

```python
sales[sales['Country'] == "United States"]
```

# CALCULATING SOME STATISTICAL MEASURES

```python
# mean of customer age
sales['Customer_Age'].mean()
```

```python
# Kernel density plot and box plot of customer age
plt.figure(1)
plt.tight_layout()

# KDE
plt.subplot(211)
sales['Customer_Age'].plot(kind='kde')

# BOXplot
plt.subplot(212)
sales['Customer_Age'].plot(kind='box' , vert=False)
plt.show()
```

```python
# mean, histogram and boxplot of order quantity
print("Mean = {}" .format(sales['Order_Quantity'].mean()))

#histogram
fig,ax = plt.subplots(nrows=2, ncols=1 , figsize=(6,6))
ax[0].hist(sales['Order_Quantity'], edgecolor='black')
ax[0].set_title("Histogram")

# BOXplot
ax[1].boxplot(sales['Order_Quantity'] , vert=False)
ax[1].set_title("BoxPlot")

plt.subplots_adjust(hspace=1)
plt.show()
```

```python
# Sales per year value counts, pie plot
print(sales['Year'].value_counts())
sales['Year'].value_counts().plot(kind='pie' , figsize=(5,5))
plt.show()
```

```python
# Sales per month value counts, bar plot
print(sales['Month'].value_counts())
sales['Month'].value_counts().plot(kind='bar' , figsize=(5,5))
plt.show()
```

```python
# Which country has the most sales quantity of sales?
print(sales['Country'].value_counts())

# barplot
sales['Country'].value_counts().plot(kind='bar')
```

```python
# Create a list of every product sold
print(sales['Product'].unique())

# bar plot of top 10 most selling items
sales['Product'].value_counts().head(10).plot(kind='bar')
plt.show()
```

```python
# Relationship between Unit_Cost and Unit_Price
plt.scatter(sales['Unit_Cost'] , sales['Unit_Price'] , alpha=0.4 , edgecolors='black')
```

```python
# Relationship between Order_Quantity and Profit
plt.scatter(sales['Order_Quantity'] , sales['Profit'] , alpha=0.4 , edgecolors='black')
```

```python
# Relationship between Profit per Country
sales[['Profit', 'Country']].boxplot(by='Country', figsize=(10,6))
```

```python
# Relationship between the Customer_Age per Country
sales[['Customer_Age', 'Country']].boxplot(by='Country', figsize=(10,6))
```

```python
# Adding a new column of Date
sales['Calculated_Date'] = sales[['Day' , 'Month' , 'Year']].apply(lambda x: '{}-{}-{}'.format(x[0] , x[1] , x[2]) , axis=1)
sales['Calculated_Date'].head(5)
```

```python
# parsing the new column in datetime object 
sales['Calculated_Date'] = pd.to_datetime(sales['Calculated_Date'])
```

```python
sales['Calculated_Date'].head()
```

```python
# Sales during the years
sales['Calculated_Date'].value_counts().plot(kind='line')
```

```python
# Increase 50 U$S revenue to every sale
sales['Revenue']+=50
```

```python
# Orders made in Canada or France
sales.loc[ (sales['Country'] == 'Canada') | ( sales['Country'] == 'France')].shape[0]
```

```python
# Bike Racks ordered from Canada
sales.loc[ (sales['Sub_Category'] == 'Bike Racks') & (sales['Country'] == 'Canada') ].shape[0]
```

```python
# Orders from each region (state) of France
sales['State'][sales['Country']=='France'].value_counts()

# Barplot of these values
sales['State'][sales['Country']=='France'].value_counts().plot(kind='bar')
```

```python
# Sales made per category
sales['Product_Category'].value_counts()

# PiePlot
sales['Product_Category'].value_counts().plot(kind='pie')
```

```python
# Gender with amount of sales
sales['Customer_Gender'].value_counts()

# BarPlot

sales['Customer_Gender'].value_counts().plot(kind='bar')
```

```python
# Sales with more than 500 in Revenue were made by men
sales.loc[ (sales['Revenue'] > 500) & (sales['Customer_Gender'] == 'M') ].shape[0]
```

```python
# top-5 sales with the highest revenue
sales.sort_values(['Revenue'] , ascending=False).head(5)
```

```python
# Mean Order_Quantity of orders with more than 10K in revenue
sales.loc[(sales['Revenue'] > 10000) , 'Order_Quantity'].mean()
```

```python
# How many orders were made in May of 2016
sales.loc[(sales['Month']  == "May") & (sales['Year']==2016) ].shape[0]
```

```python
# How many orders were made between May and July of 2016?
condition = (sales['Year']==2016) & sales['Month'].isin(['May', 'June', 'July'])
sales.loc[condition].shape[0]

# Grouped boxplot
sales[['Profit', 'Month']].boxplot(by='Month', figsize=(10,6))
```
